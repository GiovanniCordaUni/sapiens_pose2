```python
"""Esegue run_motion_analysis.py su tutti i jsonl in una cartella."""

import subprocess
from pathlib import Path

INPUT_DIR = Path(r"C:\Users\giova\OneDrive\Desktop\tesi\risultati sapiens\output_A100_1b_stabilizzato")

jsonl_files = list(INPUT_DIR.glob("**/keypoints_jsonl/*.jsonl"))
print(f"Trovati {len(jsonl_files)} file .jsonl")

for i, jsonl_file in enumerate(jsonl_files, 1):
    print(f"\n[{i}/{len(jsonl_files)}] {jsonl_file.name}")
    
    # Converti path Windows a path WSL
    wsl_input = str(jsonl_file).replace("C:\\", "/mnt/c/").replace("\\", "/")
    
    # Esegui tramite WSL con python3
    result = subprocess.run([
        "wsl", "-e", "python3",
        "/home/giovanni/sapiens_pose2/scripts/run_motion_analysis.py",
        "--input", wsl_input,
        "--auto_threshold"
    ], capture_output=True, text=True, encoding='utf-8', errors='ignore')
    
    # Mostra output
    if result.stdout:
        print(result.stdout)
    if result.returncode != 0 and result.stderr:
        print(f"ERRORE: {result.stderr}")

print("\nCompletato!")
print("Risultati in: \\\\wsl.localhost\\Ubuntu-22.04\\home\\giovanni\\sapiens_pose2\\data\\output\\test_tempi")
```
